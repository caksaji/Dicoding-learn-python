```python
# Dictionary
employee = {
  'name': 'Alice',
  'age': 30,
  'role': 'Engineer'
}
print(employee)
print(employee['name']) # Display key
print(employee.get('email')) # Display unavailable key
print(employee.get('phone', 'Not available')) # Display unavailable key with custom message
```

```python
# Transform text case
kata = 'Dicoding'
kata = kata.upper()
print(kata)
kata = kata.lower()
print(kata)
```

```python
# Trim whitespace in the beginning / end / both
print('          Dicoding          '.rstrip())
print('          Dicoding          '.lstrip())
print('          Dicoding          '.strip())
```

```python
# Trim specific string
kata = 'CodeCodeDicodingCodeCode'
print(kata.strip('Code'))
```

```python
# Check start & end of string
word = 'Dicoding Indonesia'
print(word.startswith('Dicoding'))
print(word.startswith('Dicoding'))
```

```python
# Split & join string
print(' '.join(['Dicoding', 'Indonesia', '!']))
print('123'.join(['Dicoding', 'Indonesia', '!']))

print('Dicoding Indonesia !'.split())

print('''Halo,
aku ikan,
aku suka sekali menyelam
aku tinggal di perairan.
Badanku licin dan renangku cepat.
Senang berkenalan denganmu.'''.split('\n'))
```

```python
# Replace string
string = 'Ayo belajar Coding di Dicoding'
print(string.replace('Coding', 'Pemrograman'))
```

```python
# Check string case
kata = 'DICODING'
print(kata.isupper())
print(kata.islower())
```

```python
# Check is all alphabet
print('dicoding'.isalpha())
print('dico123ding'.isalpha())
```

```python
# Check is all alphanumeric
print('dicoding'.isalnum())
print('dico123ding'.isalnum())
print('dico123ding?'.isalnum())
```

```python
# Check is all number
print('123'.isdecimal())
print('dicoding'.isdecimal())
```

```python
# Check is only space
print('     '.isspace())
print('  c  '.isspace())
```
